#Ejercicio: Uso de funciones en Python
###Ejercicio 1: Trabajar con argumentos en funciones
Los argumentos requeridos en las funciones se utilizan cuando las funciones necesitan que esos argumentos funcionen correctamente. En este ejercicio, construirás un informe de combustible que requiere información de varias ubicaciones de combustible en todo el cohete.

Comienza por crear una función que necesite tres lecturas de combustible y devuelva un informe:


```
# Función para leer 3 tanques de combustible y muestre el promedio
```

Ahora que hemos definido la función de informes, vamos a comprobarlo. Para esta misión, los tanques no están llenos:


```
# Llamamos a la función que genera el reporte print(funcion(tanque1, tanque2, tanque3))
```

En lugar de simplemente crear el informe, la función también está calculando el promedio. Mejora la legibilidad extrayendo el cálculo promedio de la función en una nueva función para que el promedio se pueda hacer de forma independiente:


```
# Función promedio 
```
Ahora actualiza la función de informes para llamando a la nueva función del promedio:


```
# Actualiza la función
```




#Solución

````python
#Creamos primero la función para leer 3 tanques de combustible
def reporte(tanque1, tanque2, tanque3):
    promedio = (tanque1 + tanque2 + tanque3) / 3
    return f"""Reporte de combustible:
    Promedio: {promedio}%
    Tanque1: {tanque1}%
    Tanque2: {tanque2}%
    Tanque3: {tanque3}% 
    """
````

````python
# Llamamos a la función que genera el reporte print(funcion(tanque1, tanque2, tanque3))
print(reporte(80, 70, 85))
````

````python
# Función promedio 
def promedio_ext(values):
    total = sum(values)
    number_of_items = len(values)
    return total / number_of_items

# Test the averaging function with a list of integers:
promedio_ext([80, 85, 81]) 
````

````python
# Actualizamos la función 
def reporte(tanque1, tanque2, tanque3):
    return f"""Reporte de combustible:
    Promedio: {promedio_ext([tanque1, tanque2, tanque3])}%
    Tanque1: {tanque1}%
    Tanque2: {tanque2}%
    Tanque3: {tanque3}% 
    """

# Llamamos a la función actualizada con otros valores
print(reporte(88, 76, 70))
````

En lugar de usar *args y **kwargs, aprovecha el hecho de que puedes usar cualquier nombre que desees. # Escribe tu nueva función

#Solución

````python
# Función con un informe preciso de la misión. Considera hora de prelanzamiento, tiempo de vuelo, destino, tanque externo y tanque interno

def reporte_mision(hora_prelanzamiento, tiempo_de_vuelo, destino, tanque_externo, tanque_interno):
    return f"""
    Mision para {destino}
    Tiempo total de vuelo: {hora_prelanzamiento + tiempo_de_vuelo} minutos
    Combustible restante: {tanque_externo + tanque_interno} gallons
    """

print(reporte_mision(14, 51, "Luna", 200000, 300000))
````

````python
# Escribimos una nueva función de reporte considerando lo anterior
def reporte_de_mision(destino, *minutos, **reserva_combustible):
    return f"""
    Mission para {destino}
    Tiempo total de vuelo: {sum(minutos)} minutos
    Combustible restante: {sum(reserva_combustible.values())}
    """

print(reporte_de_mision("Luna", 10, 15, 51, main=300000, external=200000))
````

````python
# Escribe tu nueva función
def reporte_de_mision(destino, *minutos, **reserva_combustible):
    Informe_principal = f"""
    Mision para {destino}
    Tiempo total de vuelo: {sum(minutos)} minutos
     Combustible restante: {sum(reserva_combustible.values())}
    """
    for tanques, gallones in reserva_combustible.items():
        Informe_principal += f"{tanques} tanque --> {gallones} gallones restantes\n"
    return Informe_principal

print(reporte_de_mision("Luna", 8, 11, 55, interno=300000, externo=200000))
````
