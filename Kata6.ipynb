#Solución

````python
#Lista de planetas
planetas = ['Mercurio ', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno','Urano', 'Neptuno']
````

````python
#Solicitamos el nmbre del planeta 
user_planeta = input('Inserte el nombre del planeta( comenzando con la primera letra en mayuscula) ')
````

````python
#Buscar el planeta en la lista
planeta_index = planetas.index( user_planeta)
````

````python
#Mostrtar los planetas más caercanos al sol 
print('Los planetas más cerca de' + user_planeta)
print(planetas[0:planeta_index])
````

````python
#Mostrar los planetas mas lejanos
print('Los planetas más lejos de' + user_planeta)
print(planetas[planeta_index+1:])
````
